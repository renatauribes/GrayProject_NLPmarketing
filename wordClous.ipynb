Función generadora de Word Clouds 

```python
def generate_wordclouds(data_path, description, class_column):

    # Obtener las clases únicas en el conjunto de datos

    df = pd.read_csv(data_path)

    classes = df[class_column].unique()
    
    # Generar una wordcloud para cada clase
    for class_name in classes:
      
        # Filtrar el DataFrame por la clase actual
        class_df = df[df[class_column] == class_name]
        
        # Unir el texto de las filas correspondientes a la clase
        combined_text = ' '.join(class_df[description].tolist())

        combined_text = combined_text.replace("'", "").replace("[", "").replace("]", "").replace(",", "")

        word_counts = {}
        
        # Count the occurrences of each word in the combined text
        for word in combined_text.split():
            word_counts[word] = word_counts.get(word, 0) + 1
        
        # Crear la wordcloud para la clase actual
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

        
        # Mostrar la wordcloud
        plt.figure()
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(f'Wordcloud - {class_name}')
        plt.axis('off')
        plt.show()
```

```python
generate_wordclouds(data_path, description, class_col)
```
